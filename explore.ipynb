```python
def plot(data):
    img = imread(data["image_path"])
    fig,ax = plt.subplots(1)
    x1, y1, x2, y2 = data["boxes"]
    
    # Display the image
    ax.imshow(img)
    
    # Create a Rectangle patch
    rect = patches.Rectangle((x1, y1), x2-x1, y2-y1,linewidth=1,edgecolor='r',facecolor='none')

    # Add the patch to the Axes
    ax.add_patch(rect)
    
    plt.show()
```

```python
coco_sketchy_map = pickle.load(open("../data/coco_sketchy_map.pkl", "rb"))
```

```python
query = {}
sketchy_db_path = "../data/sketchy/256x256/sketch/tx_000100000000"
for class_idx in coco_sketchy_map:
    query[class_idx] = []
    if len(coco_sketchy_map[class_idx]["sketchy"]) > 0:
        for sketch_class in coco_sketchy_map[class_idx]["sketchy"]:
            class_sketch_paths = glob.glob(os.path.join(sketchy_db_path, sketch_class, "*"))
            for sketch_path in class_sketch_paths:
                d = {"boxes": [0.0, 0.0, 256.0, 256.0], "image_path": sketch_path}
                query[class_idx].append(d)
```

```python
for class_idx in query:
    print (class_idx, len(query[class_idx]))
```
